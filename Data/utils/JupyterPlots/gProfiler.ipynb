```python
import pandas as pd
import requests
```

```python
q = pd.read_csv(r'S:\U_Proteomica\UNIDAD\software\MacrosRafa\data\Metabolomics\PESA_Integromics\Data\Proteomics\ALDH4\WorkingFiles\q2info.tsv', sep='\t')
```

```python
back = q.fid.tolist()
```

```python
g = q.fid[['apolipoprotein' in i.lower() for i in q.qdesc]].tolist()
```

```python
url = 'https://biit.cs.ut.ee/gprofiler/api/gost/profile/'

res = requests.post(url, json={
    'organism': 'mmusculus',
    'query': g,
    'sources': ["GO:MF","GO:CC","GO:BP","KEGG","REAC","WP","TF","MIRNA","CORUM","HPA","HP"],
    'user_threshold':0.1,
    'domain_scope':'custom',
    'background': back
})
```
